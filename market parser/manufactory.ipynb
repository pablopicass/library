```
import pandas as pd
import numpy as np


df_inc = pd.read_csv('/indicators.csv')
df_inr = pd.read_csv('/industry.csv')
df_pfs = pd.read_csv('/pfs.csv')

df = pd.merge(left=df_inc, right=df_pfs, on='ticker')
df = pd.merge(left=df, right=df_inr, on='ticker')

df.replace(to_replace='NaN', value=np.nan, inplace=True)
```

```
def norm(df, sector=None, industry=None):
  if sector:
    df_sect = pd.DataFrame(df[df['industry'] == sector])
  if industry:
    df_sect = pd.DataFrame(df[df['sector'] == industry])
  
  cols = ['recomendation',	'P/E',	'future P/E',	'PEG',	'Debt/Equity',	'ROE',	'shorts, %',	'RSI',	'SMA50 - SMA200, %',	'forecast FINVIZ, %', 'pfs']

  for col in cols:
    mean = round(df_sect[col].mean(), 2)
    df_sect[col].fillna(mean, inplace=True)
    df_sect[col] = (df_sect[col] - min(df_sect[col])) / (max(df_sect[col]) - min(df_sect[col]))
  
  return df_sect


def negative(df):
  cols = ['recomendation',	'P/E',	'future P/E',	'PEG',	'Debt/Equity', 'shorts, %',	'RSI']

  for col in cols:
    df[col] = 1 - df[col]
  
  return df
```

```
df_sect = norm(df, sector='Consumer Cyclical')
df_sect = negative(df_sect)

df_sect['platinum score'] = df_sect['recomendation'] + df_sect['P/E'] + df_sect['future P/E'] + df_sect['PEG'] + df_sect['Debt/Equity'] + df_sect['ROE'] + df_sect['shorts, %'] + df_sect['RSI'] + df_sect['SMA50 - SMA200, %'] + df_sect['forecast FINVIZ, %'] + df_sect['pfs']
```

```
df_sect.sort_values(by='platinum score', ascending=False, inplace=True)
df_sect.head(30)
```
